```python
import os
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn import metrics, svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import precision_recall_fscore_support
```

```python
file_path = '/home/anagha/Documents/Semester7/LP-II/DMW/MiniProject/pima-indians-diabetes.csv'

dataset = pd.read_csv(file_path, names = ['No. of pregnancies', 'Plasma Glucose Conc.', 'Diastolic BP', 'Tricep skin fold thickness', '2 hr Serum insulin', 'BMI', 'Diabetes pedigree function', 'Age', 'P/N'])
```

```python
# Data has no null values. 
# But holds a value of 0 for many features that do not contain 0 in their range of possible values.

# Columns that cannot hold the value 0, but are present in the dataset:

# 1. Plasma Glucose Conc.
# 2. Diastolic BP
# 3. Tricep skin fold thickness
# 4. 2 hr Serum insulin
# 5. BMI
# 6. Diabetes pedigree function

# We need to handle these values appropriately to get the best estimate.
# (Consider these zeroes to be null values.)

# Limitations:
# 1. Small dataset
# 2. Represents a small part of the demographic.
# 3. Data was collected during the late 1900s, making it outdated.

# Since we have a small dataset, we should avoid dropping records where we have missing values.
# We will replace the missing values with the average vof all the values in that column.
```

```python
# Replacing all the missing values with averages

mis = ['Plasma Glucose Conc.', 'Diastolic BP', 'Tricep skin fold thickness', '2 hr Serum insulin', 'BMI', 'Diabetes pedigree function']

averages = pd.DataFrame({'feature' : mis, 'positive' : np.repeat(0, len(mis)), 'negative' : np.repeat(0, len(mis))})

i = 0
for col in mis:
    pos = 0
    neg = 0
    pos_count = 0
    neg_count = 0
    for ind in dataset.index:
        if dataset.loc[ind, 'P/N'] == 1:
            pos += dataset.loc[ind, col]
            pos_count += 1
        else:
            neg += dataset.loc[ind, col]
            neg_count += 1
    averages.iloc[i, 1] = pos / pos_count
    averages.iloc[i, 2] = neg / neg_count
    i += 1

print(averages.head())

for ind in dataset.index:
    for col in mis:
        if dataset.loc[ind, col] == 0:
            if dataset.loc[ind, 'P/N'] == 1:
                dataset.loc[ind, [col]] = float(averages.loc[averages['feature'] == col]['positive'])
            else:
                dataset.loc[ind, [col]] = float(averages.loc[averages['feature'] == col]['negative'])

# for ind in dataset.index:
#     for col in mis:
#         if dataset[col][ind] == 0:
#             dataset.loc[ind, 'Missing'] = 1
#             break
#         else:
#             dataset.loc[ind, 'Missing'] = 0
```

```python
# We can use discretization for the 'No. of pregnancies', 'BMI' and 'Age' columns.
# Equal Width Discretizatoin can be used, since value spread is not of importance.
```

```python
# Equal Width Discretization

# Discretize the 'Age' column
discretizer = KBinsDiscretizer(n_bins=12, encode='ordinal', strategy='uniform')

p = np.array(dataset['Age']).reshape(-1, 1)
age_disc = discretizer.fit_transform(p)
age_disc = age_disc.flatten()
dataset['Age'] = age_disc

# Discretize the 'No. of pregnancies' column
discretizer = KBinsDiscretizer(n_bins=9, encode='ordinal', strategy='uniform')

p = np.array(dataset['No. of pregnancies']).reshape(-1, 1)
age_disc = discretizer.fit_transform(p)
age_disc = age_disc.flatten()
dataset['No. of pregnancies'] = age_disc

# # Discretize the 'BMI' column
# discretizer = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')

# p = np.array(dataset['BMI']).reshape(-1, 1)
# age_disc = discretizer.fit_transform(p)
# age_disc = age_disc.flatten()
# dataset['BMI'] = age_disc
```

```python
# We will perform Stratified K Fold Cross Validation, since we are performing binary classification.

X = dataset.iloc[:, 0:-1]
y = dataset.iloc[:, -1]

skf = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 1)

for train_index, test_index in skf.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
```

```python
# MODEL I

# Logistic Regression

log_reg_acc = []
log_reg_prec = []
log_reg_rec = []

def logistic_regression(x_train, y_train, x_test, y_test):
    logisticRegr = LogisticRegression(solver = 'newton-cg')

    logisticRegr.fit(x_train, y_train)
    predictions = logisticRegr.predict(x_test)
    log_reg_acc.append(logisticRegr.score(x_test, y_test))
    cm = metrics.confusion_matrix(y_test, predictions)
    confusion_analysis = precision_recall_fscore_support(y_true = y_test, y_pred = predictions, average = 'micro')
    log_reg_prec.append(confusion_analysis[0])
    log_reg_rec.append(confusion_analysis[1])
    print('Confusion Matrix: \n',cm)
    

for train_index, text_index in skf.split(X, y):
    train_X = X.loc[train_index]
    train_y = y.loc[train_index]
    test_X = X.loc[test_index]
    test_y = y.loc[test_index]
    logistic_regression(train_X, train_y, test_X, test_y)
    
print("Logistic Regression Accuracies for Stratifies K Fold Cross Validation:")
print(log_reg_acc)
print("Mean Accuracy:", sum(log_reg_acc) / len(log_reg_acc))
print("Mean Precision:", sum(log_reg_prec) / len(log_reg_prec))
print("Mean Recall:", sum(log_reg_rec) / len(log_reg_rec))
```

```python
# MODEL II

# k-Nearest Neighbours

k_near_acc = []
k_near_prec = []
k_near_rec = []

def knearest(x_train, y_train, x_test, y_test):
    model = KNeighborsClassifier(n_neighbors=3)
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    k_near_acc.append(metrics.accuracy_score(y_test, predictions))
    cm = metrics.confusion_matrix(y_test, predictions)
    confusion_analysis = precision_recall_fscore_support(y_true = y_test, y_pred = predictions, average = 'micro')
    k_near_prec.append(confusion_analysis[0])
    k_near_rec.append(confusion_analysis[1])
    print('Confusion Matrix: \n',cm)
    
for train_index, text_index in skf.split(X, y):
    train_X = X.loc[train_index]
    train_y = y.loc[train_index]
    test_X = X.loc[test_index]
    test_y = y.loc[test_index]
    knearest(train_X, train_y, test_X, test_y)

print("k Nearest Neighbours Accuracies for Stratifies K Fold Cross Validation:")
print(k_near_acc)
print("Mean Accuracy:", sum(k_near_acc) / len(k_near_acc))
print("Mean Precision:", sum(k_near_prec) / len(k_near_prec))
print("Mean Recall:", sum(k_near_rec) / len(k_near_rec))
```

```python
# MODEL III

# Naive Bayes

naive_bayes_acc = []
naive_bayes_prec = []
naive_bayes_rec = []

def naive_bayes(x_train, y_train, x_test, y_test):
    model = GaussianNB()
    model.fit(x_train, y_train)
    predictions = model.predict(x_test)
    naive_bayes_acc.append(metrics.accuracy_score(y_test, predictions))
    cm = metrics.confusion_matrix(y_test, predictions)
    confusion_analysis = precision_recall_fscore_support(y_true = y_test, y_pred = predictions, average = 'micro')
    naive_bayes_prec.append(confusion_analysis[0])
    naive_bayes_rec.append(confusion_analysis[1])
    print('Confusion matrix: \n', cm)

for train_index, text_index in skf.split(X, y):
    train_X = X.loc[train_index]
    train_y = y.loc[train_index]
    test_X = X.loc[test_index]
    test_y = y.loc[test_index]
    naive_bayes(train_X, train_y, test_X, test_y)

print("Naive Bayes Accuracies for Stratifies K Fold Cross Validation:")
print(naive_bayes_acc)
print("Mean Accuracy:", sum(naive_bayes_acc) / len(naive_bayes_acc))
print("Mean Precision:", sum(naive_bayes_prec) / len(naive_bayes_prec))
print("Mean Recall:", sum(naive_bayes_rec) / len(naive_bayes_rec))
```

```python
# MODEL IV

# Support Vector Machines

svm_acc = []
svm_prec = []
svm_rec = []

def svm_classifier(x_train, y_train, x_test, y_test):
    clf = svm.SVC(kernel='linear')
    clf.fit(x_train, y_train)
    predictions = clf.predict(x_test)
    svm_acc.append(metrics.accuracy_score(y_test, predictions))
    cm = metrics.confusion_matrix(y_test, predictions)
    confusion_analysis = precision_recall_fscore_support(y_true = y_test, y_pred = predictions, average = 'micro')
    svm_prec.append(confusion_analysis[0])
    svm_rec.append(confusion_analysis[1])
    print('Confusion matrix: \n', cm)

for train_index, text_index in skf.split(X, y):
    train_X = X.loc[train_index]
    train_y = y.loc[train_index]
    test_X = X.loc[test_index]
    test_y = y.loc[test_index]
    svm_classifier(train_X, train_y, test_X, test_y)

print("Support Vector Machine Accuracies for Stratifies K Fold Cross Validation:")
print(svm_acc)
print("Mean Accuracy:", sum(svm_acc) / len(svm_acc))
print("Mean Precision:", sum(svm_prec) / len(svm_prec))
print("Mean Recall:", sum(svm_rec) / len(svm_rec))
```
